```python
import pandas as pd

ads = pd.read_csv("../Data/Social_Network_Ads.csv")

ads.head()
```

```python
import seaborn as sns

sns.barplot(data=ads, x="Purchased", y="EstimatedSalary")
```

```python
from sklearn.model_selection import train_test_split

X = pd.get_dummies(ads[["Gender", "Age","EstimatedSalary"]], drop_first=True)
y = ads["Purchased"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2023)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier() 

dt = dt.fit(X_train, y_train)

print(f"Train Accuracy: {round(dt.score(X_train, y_train), 3)}")
print(f"Test Accuracy: {round(dt.score(X_test, y_test), 3)}")
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(max_depth=3, min_samples_leaf=10) 

dt = dt.fit(X_train, y_train)

print(f"Train Accuracy: {round(dt.score(X_train, y_train), 3)}")
print(f"Test Accuracy: {round(dt.score(X_test, y_test), 3)}")
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(
    criterion="entropy",
    max_depth=None, 
    min_samples_leaf=None,
    class_weights=None
) 
```
